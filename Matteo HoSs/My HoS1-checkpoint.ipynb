```python
import numpy as np
from astropy.io import fits
from astropy.table import Table
import matplotlib.pyplot as plt
import os
```

```python
tab = Table.read('SMF_Baldry2012.txt', format='ascii.commented_header')
Mstar = tab['Mass']
Ndens = tab['Density']
```

```python
# Plot della SMF
plt.figure(figsize=(6,4))
plt.semilogy(Mstar, Ndens, marker='o', linestyle='-')
plt.xlabel(r'$\log_{10}(M_*/M_\odot)$')
plt.ylabel(r'$\phi\ (Mpc^{-3}\ dex^{-1})$')
plt.title('Stellar Mass Function')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Parametri cosmologici / simulazione
h0 = 0.7
VOLUME = 150.0 ** 3  # Mpc^3
BIN_WIDTH = 0.2  # dex (dato dal problema)

n_expected = Ndens * VOLUME * BIN_WIDTH  # numero medio atteso in ciascun bin
# campioniamo il numero di oggetti effettivi per bin con processo Poisson
n_in_bin = np.random.poisson(n_expected)

# Generiamo un array di M* campionato uniformemente dentro ogni bin
sampled_Mstar = []
half = BIN_WIDTH / 2.0
for i, n in enumerate(n_in_bin):
    if n <= 0:
        continue
    low = Mstar[i] - half
    high = Mstar[i] + half
    sampled = np.random.uniform(low=low, high=high, size=n)
    sampled_Mstar.append(sampled)

if len(sampled_Mstar) == 0:
    raise RuntimeError('Nessuna galassia campionata: controlla SMF o il volume.')

sampled_Mstar = np.concatenate(sampled_Mstar)
print(f"Numero di galassie campionate dalla SMF: {len(sampled_Mstar)}")

# Istogramma (numero di galassie campionate per bin, controllo)
plt.figure(figsize=(6,4))
plt.hist(sampled_Mstar, bins=np.append(Mstar - half, Mstar[-1] + half))
plt.xlabel(r'$\log_{10}(M_*/M_\odot)$')
plt.ylabel('Numero campionato (per bin)')
plt.title('Controllo: istogramma galassie campionate dalla SMF')
plt.tight_layout(); plt.show()
```

```python
halo_file = 'MR7_z0.0_L150_h0.7.fits'
sim = Table.read(halo_file)
print('Colonne halo:', sim.colnames)

MVIR = np.array(sim['MVIR'], dtype=float)
logMh = np.log10(MVIR)

# ordini decrescenti
halo_sort_idx = np.argsort(logMh)[::-1]
gal_sort_idx = np.argsort(sampled_Mstar)[::-1]

n_match = min(len(halo_sort_idx), len(gal_sort_idx))
print(f"Matcheremo {n_match} oggetti (min tra #halo e #gal).")

matched_logMh = logMh[halo_sort_idx[:n_match]]
matched_Mstar = sampled_Mstar[gal_sort_idx[:n_match]]

# Plot Mh vs M* (entrambi in log10(Msun)). Ogni punto  un halo + gal abbinata.
plt.figure(figsize=(6,5))
plt.scatter(matched_logMh, matched_Mstar, s=6, alpha=0.5)
plt.xlabel(r'$\log_{10}(M_{halo}/M_\odot)$ (MVIR)')
plt.ylabel(r'$\log_{10}(M_*/M_\odot)$')
plt.title('Abundance matching: MVIR (halo) vs M* (galaxia)')
plt.grid(alpha=0.3)
plt.tight_layout(); plt.show()
```

```python
Omega_b = 0.045
Omega_m = 0.31
Omega_DM = Omega_m - Omega_b
fb_Omega_b_over_Omega_DM = Omega_b / Omega_DM
fb_Omega_b_over_Omega_m = Omega_b / Omega_m
print('Omega_DM =', Omega_DM)
print('Omega_b / Omega_DM =', fb_Omega_b_over_Omega_DM)
print('Omega_b / Omega_m   =', fb_Omega_b_over_Omega_m)

# rapporto M*/Mhalo in unit lineare
mass_ratio = 10 ** (matched_Mstar - matched_logMh)

plt.figure(figsize=(6,5))
plt.scatter(matched_logMh, mass_ratio, s=6, alpha=0.5)
plt.yscale('log')
plt.xlabel(r'$\log_{10}(M_{halo}/M_\odot)$')
plt.ylabel(r'$M_*/M_{halo}$ (lineare)')
plt.title('Mhalo vs (M*/Mhalo) dalla SHAM')
plt.grid(alpha=0.1)
plt.tight_layout(); plt.show()
```

```python
tng_file = 'TNG_Mhalo_Mstar_h0.7.fits'
if not os.path.exists(tng_file):
    print(f"Attenzione: non trovo {tng_file}. Salta il confronto con TNG se non  presente." )
else:
    tng = Table.read(tng_file)
    print('Colonne TNG:', tng.colnames)
    # Cerchiamo colonne plausibili
    possible_Mh = [c for c in tng.colnames if 'halo' in c.lower() or 'mhalo' in c.lower()]
    possible_Ms = [c for c in tng.colnames if 'star' in c.lower() or 'mstar' in c.lower()]
    if len(possible_Mh) == 0 or len(possible_Ms) == 0:
        raise KeyError('Non trovo colonne chiare per Mhalo o Mstar nel file TNG. Controlla tabelle.')

    col_Mh = possible_Mh[0]
    col_Ms = possible_Ms[0]
    print('Uso per TNG:', col_Mh, col_Ms)

    tng_logMh = np.array(np.log10(tng[col_Mh]), dtype=float)
    tng_logMs = np.array(np.log10(tng[col_Ms]), dtype=float)

    # calcoliamo mediana di M*/Mhalo in bin di Mhalo
    bins = np.arange(10.0, 15.0, 0.25)
    inds = np.digitize(tng_logMh, bins)
    bin_centers = bins[:-1] + 0.25/2
    median_ratio = []
    for i in range(1, len(bins)):
        sel = (inds == i)
        if sel.sum() == 0:
            median_ratio.append(np.nan)
        else:
            mr = np.median(10 ** (tng_logMs[sel] - tng_logMh[sel]))
            median_ratio.append(mr)
    median_ratio = np.array(median_ratio)

    # Plot confronto: SHAM scatter + TNG mediana
    plt.figure(figsize=(6,5))
    plt.scatter(matched_logMh, mass_ratio, s=6, alpha=0.3, label='SHAM (this work)')
    plt.plot(bin_centers, median_ratio, marker='o', linewidth=2, label='TNG median')
    plt.yscale('log')
    plt.xlabel(r'$\log_{10}(M_{halo}/M_\odot)$')
    plt.ylabel(r'$M_*/M_{halo}$')
    plt.title('Confronto M*/Mhalo: SHAM vs TNG (mediana)')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout(); plt.show()
```

```python
halo_file = 'MR7_z0.0_L150_h0.7.fits'
if not os.path.exists(halo_file):
    raise FileNotFoundError(f"Non trovo {halo_file}. Metti il file nella cartella e rilancia.")

sim = Table.read(halo_file)
print('Colonne halo:', sim.colnames)

# alcuni controlli: colonne di interesse sono MVIR, TYPE, INFALLVMAX (se presente)
if 'MVIR' not in sim.colnames:
    raise KeyError('Il catalogo degli haloes non contiene la colonna "MVIR".')

MVIR = np.array(sim['MVIR'], dtype=float)
# evitiamo valori non fisici
mask_positive = MVIR > 0
MVIR = MVIR[mask_positive]
# manteniamo indici mapping per ritrovare TYPE ecc. nel caso serva
sim_masked = sim[mask_positive]

if 'INFALLVMAX' in sim_masked.colnames:
    vmax = np.array(sim_masked['INFALLVMAX'], dtype=float)
    # attenzione a zero o valori non fisici
    mask_pos_v = vmax > 0
    # prenderemo solo haloes con vmax>0 per il ranking alternativo
    logv = np.log10(vmax[mask_pos_v])
    idx_v_sort = np.argsort(logv)[::-1]

    # numero di match come prima
    n_match_v = min(len(idx_v_sort), len(gal_sort_idx))
    matched_idx_v = idx_v_sort[:n_match_v]

    matched_logMh_v = logMh[mask_pos_v][matched_idx_v]
    matched_Mstar_v = sampled_Mstar[gal_sort_idx[:n_match_v]]

    # plot confronto SHAM (MVIR-ranking) vs SHAM (Vmax-ranking)
    plt.figure(figsize=(6,5))
    plt.scatter(matched_logMh[:1000], matched_Mstar[:1000], s=6, alpha=0.3, label='MVIR-based (primi 1000)')
    plt.scatter(matched_logMh_v[:1000], matched_Mstar_v[:1000], s=6, alpha=0.3, label='INFALLVMAX-based (primi 1000)')
    plt.xlabel(r'$\log_{10}(M_{halo}/M_\odot)$')
    plt.ylabel(r'$\log_{10}(M_*/M_\odot)$')
    plt.title('Confronto ranking: MVIR vs INFALLVMAX')
    plt.legend(); plt.grid(alpha=0.3)
    plt.tight_layout(); plt.show()

    # Nota: l'uso di INFALLVMAX tende a spostare verso l'alto la posizione dei satelliti
    # pi massicci che hanno perso massa; l'effetto globale dipende dalla frazione di
    # satelliti e dalla loro storia di stripping.
else:
    print('Colonna INFALLVMAX non presente: salto questa parte.')
```

```python
# ==========================
# SEMPLICE ANALISI SHAM
# ==========================
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
import os

# --------------------------
# 1. Leggiamo la Stellar Mass Function (SMF)
# --------------------------
tab = Table.read('SMF_Baldry2012.txt', format='ascii.commented_header')
Mstar_bins = tab['Mass']     # log10(M*/Msun)
Ndens = tab['Density']       # densit [Mpc^-3 dex^-1]

# Plot della SMF
plt.figure(figsize=(6,4))
plt.semilogy(Mstar_bins, Ndens, marker='o')
plt.xlabel(r'$\log_{10}(M_*/M_\odot)$')
plt.ylabel('Densit [Mpc$^{-3}$ dex$^{-1}$]')
plt.title('Stellar Mass Function')
plt.grid(True)
plt.tight_layout()
plt.show()

# --------------------------
# 2. Convertiamo la SMF in numero di galassie attese
# --------------------------
VOLUME = 150.0 ** 3   # volume della simulazione in Mpc^3
BIN_WIDTH = 0.2       # larghezza dei bin in dex

# Numero atteso = densit  volume  larghezza bin
n_gal_per_bin = (Ndens * VOLUME * BIN_WIDTH).astype(int)

# --------------------------
# 3. Campioniamo masse stellari in modo uniforme nei bin
# --------------------------
sampled_logMstar = []
half_bin = BIN_WIDTH / 2.0

for i, n in enumerate(n_gal_per_bin):
    if n <= 0:
        continue
    low = Mstar_bins[i] - half_bin
    high = Mstar_bins[i] + half_bin
    sampled_logMstar.append(np.random.uniform(low, high, size=n))

# Uniamo tutto in un singolo array
sampled_logMstar = np.concatenate(sampled_logMstar)
print("Numero totale di galassie campionate:", len(sampled_logMstar))

# Istogramma di controllo
plt.figure(figsize=(6,4))
plt.hist(sampled_logMstar, bins=np.append(Mstar_bins - half_bin, Mstar_bins[-1] + half_bin))
plt.xlabel(r'$\log_{10}(M_*/M_\odot)$')
plt.ylabel('Numero di galassie')
plt.title('Galassie campionate dalla SMF')
plt.tight_layout()
plt.show()

# --------------------------
# 4. Leggiamo i dati degli aloni dalla simulazione Millennium
# --------------------------
halo_file = 'MR7_z0.0_L150_h0.7.fits'
if not os.path.exists(halo_file):
    raise FileNotFoundError(f"Non trovo {halo_file}. Metti il file nella cartella e rilancia.")

sim = Table.read(halo_file)
print("Colonne disponibili nel catalogo:", sim.colnames)

# Prendiamo la massa degli aloni
Mhalo = np.array(sim['MVIR'], dtype=float)
logMhalo = np.log10(Mhalo)

# --------------------------
# 5. Applichiamo il Sub-Halo Abundance Matching (SHAM)
#    --> ordiniamo galassie e aloni per massa e abbiniamo
# --------------------------
# Ordine decrescente per aloni e galassie
halo_sorted_idx = np.argsort(logMhalo)[::-1]
gal_sorted_idx = np.argsort(sampled_logMstar)[::-1]

n_match = min(len(halo_sorted_idx), len(gal_sorted_idx))
print("Numero di oggetti abbinati:", n_match)

matched_logMhalo = logMhalo[halo_sorted_idx[:n_match]]
matched_logMstar = sampled_logMstar[gal_sorted_idx[:n_match]]

# --------------------------
# 6. Plot Mhalo vs M*
# --------------------------
plt.figure(figsize=(6,5))
plt.scatter(matched_logMhalo, matched_logMstar, s=5, alpha=0.4)
plt.xlabel(r'$\log_{10}(M_{halo}/M_\odot)$')
plt.ylabel(r'$\log_{10}(M_*/M_\odot)$')
plt.title('Abundance Matching: Mhalo vs M*')
plt.grid(True)
plt.tight_layout()
plt.show()

# --------------------------
# 7. Rapporto M*/Mhalo e confronto con frazione barionica cosmica
# --------------------------
Omega_b = 0.045
Omega_m = 0.31
Omega_DM = Omega_m - Omega_b
fb = Omega_b / Omega_DM
print("Frazione barionica cosmica (b/DM):", fb)

mass_ratio = 10 ** (matched_logMstar - matched_logMhalo)

plt.figure(figsize=(6,5))
plt.scatter(matched_logMhalo, mass_ratio, s=5, alpha=0.4)
plt.axhline(fb, color='red', linestyle='--', label='b/DM')
plt.yscale('log')
plt.xlabel(r'$\log_{10}(M_{halo}/M_\odot)$')
plt.ylabel(r'$M_*/M_{halo}$')
plt.title('Rapporto M*/Mhalo')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
